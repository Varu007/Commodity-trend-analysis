Import Libraries

```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Fetch Financial Data

```python
# Fetch data for Gold over the last 1 year
gold = yf.download('GC=F', start='2023-07-01', end='2024-07-01')
gold.head()
```

Plot Price Trend

```python
plt.figure(figsize=(12,6))
plt.plot(gold['Close'], label='Gold Price', color='gold')
plt.title('Gold Price Trend (Last 1 Year)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

Comparison (Gold vs Crude Oil)

```python
crude = yf.download('CL=F', start='2023-07-01', end='2024-07-01')

plt.figure(figsize=(12,6))
plt.plot(gold['Close'], label='Gold', color='gold')
plt.plot(crude['Close'], label='Crude Oil', color='black')
plt.title('Gold vs Crude Oil Price Trend', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

Normalize Prices

```python
gold_norm = gold['Close'] / gold['Close'].iloc[0]
crude_norm = crude['Close'] / crude['Close'].iloc[0]

plt.figure(figsize=(12,6))
plt.plot(gold_norm, label='Gold (Normalized)', color='gold')
plt.plot(crude_norm, label='Crude Oil (Normalized)', color='black')
plt.title('Normalized Price Comparison: Gold vs Crude Oil', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.legend()
plt.grid(True)
plt.show()
```
